```python
import pandas as pd
import sqlite3
```

## Connection to db

```python
db_con = sqlite3.connect('../data/checking-logs.sqlite.sqlite')
```

##  Analyze only the users and not the admins

```python
table = pd.io.sql.read_sql("select timestamp from checker WHERE uid LIKE 'user_%'", db_con, parse_dates='timestamp')
table['date'] = table['timestamp'].dt.date
table['hour'] = table['timestamp'].dt.hour
table['dayofweek'] = table['timestamp'].dt.dayofweek
table
```

## For each hour, calculate the average number of commits on working days and on weekends (if there were no commits in an hour, do not use it to calculate the average) use these values for your graph, for example: Mon, 17-18: 5 commits, Tue, 17-18: 6 commits, Wed, 17-18: 7 commits

```python
table['day'] = table['dayofweek'].apply(lambda x: 'weekend' if x > 4 else 'working_day')
table
```

```python
tab = table.groupby(['day', 'date', 'hour']).count()
table = tab.groupby(['day', 'hour']).mean().reset_index()
table
```

```python
data = pd.pivot(table, values='dayofweek', index='hour', columns='day')
data
```

## Make a graph

```python
data.plot(kind='bar', subplots=True, fontsize=8, figsize=(15,8),
    title='Commits per day', color=['xkcd:plum', 'xkcd:sienna'])
```
